# Feature Engineering for ML models

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sales.csv")
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
```

```python
df.head()
```

```python
# Rolling mean — captures recent trend/level
df["roll_mean_7"] = df["sales"].rolling(window=7).mean()
df["roll_mean_30"] = df["sales"].rolling(window=30).mean()
```

```python
df.head(15)
```

```python
# Rolling std — captures recent volatility
df["roll_std_7"] = df["sales"].rolling(window=7).std()
```

```python
df.head(15)
```

```python
# Rolling min and max — capture recent range
df['roll_min_7']   = df['sales'].rolling(window=7).min()
df['roll_max_7']   = df['sales'].rolling(window=7).max()
```

```python
df.head(15)
```

```python
# Rolling median — robust to outliers, unlike mean
df['roll_median_7'] = df['sales'].rolling(window=7).median()
```
