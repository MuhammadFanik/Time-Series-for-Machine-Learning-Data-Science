# Setting a Time Index in Pandas

```python
df = pd.read_csv("D:/Complete ML/Time Series Data/1-Time-series-analysis/sales.csv")
df.head()
```

```python
df.shape
```

```python
# Convert the date column to datetime and set as index
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
```

```python
df.head()
```

# Resampling - Changing Frequency

```python
# Downsampling (going to lower frequency)
monthly_sales_total = df["sales"].resample("MS").sum()    # Total sales per month
monthly_sales_avg = df["sales"].resample("MS").mean()     # Monthly average of sales

print(monthly_sales_total)
print("-"*40)
print(monthly_sales_avg)
```

```python
df.head()
```

```python
# Upsampling (going to higher frequency)
daily1 = df["sales"].resample("D").ffill()  # Forward fill
daily2 = df["sales"].resample("D").interpolate()    # Linear Interpolation

print(daily1)
print("-" * 40)
print(daily2)
```

### Forward Fill
##### The idea: "I dont know what happened after this point, so I'll assume it stayed the same until the next known value". It takes the last known value and carries it forward to fill all the gaps

```python
df.head()
```

```python
# Step 1: Downsample to monthly (sum of sales per month)
monthly = df["sales"].resample("MS").sum()
print("First 6 rows of monthly data")
print("-" * 40)
print(monthly.head(6))
```

```python
# Step 2: Upsample back to daily using forward fill
daily_forward_fill = monthly.resample("D").ffill()
print("After ffill to daily")
print("-" * 40)
print(daily_forward_fill.head(35))
```

### Interpolation
##### The idea: "I know the value at point A and point B, so I will draw a straight line between them and estimate values in between". It fills gaps by calculating what the value would be if it changed smoothly and linearly from one known point to the next

```python
monthly_sales_total
```

```python
# Now fill this with interpolation
daily_interpolation = monthly.resample("D").interpolate()
print("After interpolation to daily")
print("-" * 40)
print(daily_interpolation.head(35))
```
